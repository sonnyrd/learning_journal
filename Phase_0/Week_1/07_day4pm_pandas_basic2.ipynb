## Query dataset

```python
#Multiple

df[
    (df['year_id']==2001) &
    (df['fran_id']== 'Celtics') &
    (df['game_location']=='H') &
    (df['game_result']== 'W')
] [['fran_id','pts','opp_fran','opp_pts']]
```

```python
df.query("year_id == 2001 & fran_id == 'Celtics' & game_location == 'H' & game_result == 'W' ")
```

```python
df.query(
    "year_id == 2001 &\
    fran_id == 'Celtics' &\
    game_location == 'H' &\
    game_result == 'W' ")
```

```python
df.query(
    "year_id == 2001 &\
    fran_id == 'Celtics' &\
    game_location == 'H' &\
    result == 'W' ")\
    [['fran_id','pts','opp_fran','opp_pts']]
```

### Grouping and Aggregating

```python
df.sample(5)
```

```python
#mengelompokan data

gr = df.groupby('game_result')
gr
```

```python
#simulasi saja untuk menampilkan

gr.get_group('L')
```

```python
#aggregating
df.groupby('game_result')['pts'].mean()

#round(df.groupby('game_result')['pts'].mean(),2)
#jika ingin dituliskan dengan desimal tidak lebih dari
```

rata-rata pts untuk group team yang kalah(L) adalah 97.34, sedangkan untuk rata-rata pts group yang menang adalah 108.11

```python
#kombinasi group by dan query
# df.groupby('A').agg({'B': ['min', 'max'], 'C': 'sum'})

df[
    (df['year_id']>2010)&
    (df['fran_id']=='Spurs')
].groupby(['year_id','result']).agg({'game_id':'count'})
```
