#### Feature Selection

```python
# Display train-set
X_train
```

karena `Name` tidak ada hubungan dengan `Rating` maka dapat diasumikan bahwa column ini dihilangkan. Hal ini dibuktikan dengan nama `Danu Ronaldo` tidak ada kaitanya dengan `Cristiano ROnaldo` sehingga tidak ada pengaruh apapun terhadap `Rating`.

##### Feature Scaling / Encoding
Train : Numeric : Feature Scaling / Category : Feature Encoding, digabung menjadi train_set --> model_train \
Test : Numeric : Feature Scaling / Category : Feature Encoding, digabung menjadi test --> predict

##### Split Numerical columns and Categorical Column

```python
num_columns = X_train.select_dtypes(include=np.number).columns.tolist()
cat_columns = X_train.select_dtypes(include=['object']).columns.tolist()

print('Numeric Columns: ',num_columns)
print('Categorical Columns :',cat_columns)
```

```python
X_train_num = X_train[num_columns]
X_train_cat = X_train[cat_columns]

X_test_num = X_test[num_columns]
X_test_cat = X_test[cat_columns]
```

##### Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train_num) # fit cuma boleh di train

X_train_num_scale = scaler.transform(X_train_num)
X_test_num_scale = scaler.transform(X_test_num)
X_test_num_scale
```

##### Feature Encoding

```python
#ordinal Encoder merubah dari string ke numeric dimana ada tingkatan : SD, SMP, SMA (diurutkan berdasarkan abjad)
#OneHot Encoder Merubah dari string ke numeric tidak ada tingkatan : JKT, BDG, SBY

from sklearn.preprocessing import OrdinalEncoder #label encoder / onehot encoder
encoder = OrdinalEncoder()
encoder.fit(X_train_cat) # fit cuma boleh di train

X_train_cat_enc = encoder.transform(X_train_cat)
X_test_cat_enc = encoder.transform(X_test_cat)
X_train_cat_enc
```

```python
X_train_cat
```

### Concat between Numerical Columns and Categorical Columns

```python
X_train_final = np.concatenate([X_train_num_scale,X_train_cat_enc],axis=1)
X_test_final = np.concatenate([X_test_num_scale,X_test_cat_enc],axis=1)

X_train_final
```

```python
X_train_final_df = pd.DataFrame(X_train_final, columns=[num_columns + cat_columns])
X_train_final_df
```

### vi. Model Definition


    Bagian ini berisi cell untuk mendefinisikan model. Jelaskan alasan menggunakan suatu algoritma/model, hyperparameter yang dipakai, jenis penggunaan metrics yang dipakai, dan hal lain yang terkait dengan model.


```python
from sklearn.linear_model import LinearRegression
model_lin_reg = LinearRegression()
```

### vii. Model Training


    Cell pada bagian ini hanya berisi code untuk melatih model dan output yang dihasilkan. Lakukan beberapa kali proses training dengan hyperparameter yang berbeda untuk melihat hasil yang didapatkan. Analisis dan narasikan hasil ini pada bagian Model Evaluation.


```python
model_lin_reg.fit(X_train_final,y_train)
```

### Viii. Model Evaluation


    Pada bagian ini, dilakukan evaluasi model yang harus menunjukkan bagaimana performa model berdasarkan metrics yang dipilih. Hal ini harus dibuktikan dengan visualisasi tren performa dan/atau tingkat kesalahan model. Lakukan analisis terkait dengan hasil pada model dan tuliskan hasil analisisnya.


```python
y_pred_train = model_lin_reg.predict(X_train_final)
y_pred_test = model_lin_reg.predict(X_test_final)
y_pred_test.shape
```

```python
from sklearn.metrics import mean_absolute_error

mae_train = mean_absolute_error(y_train,y_pred_train)
mae_test = mean_absolute_error(y_test,y_pred_test)

print('Error Train :', mae_train)
print('Error Test :', mae_test)
```
